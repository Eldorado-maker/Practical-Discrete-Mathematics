Load any creative images from wikipedia. Enlarge the image double of its size.

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the image (replace 'your_image.jpg' with your uploaded file name)
image_path = "Kamen_Rider_Decade_logo.jpg"
image = Image.open(image_path)

# Display the original image
print("Original image size:", image.size)
plt.figure(figsize=(6, 6))
plt.title("Original Image")
plt.imshow(image)
plt.axis('off')
plt.show()

# Resize the image to double its size
new_size = (image.size[0] * 2, image.size[1] * 2)
resized_image = image.resize(new_size)

# Save and display the resized image
resized_image.save("resized_image.jpg")
print("Resized image size:", resized_image.size)
plt.figure(figsize=(8, 8))
plt.title("Resized Image (2x)")
plt.imshow(resized_image)
plt.axis('off')
plt.show()

# Transpose (rotate) the resized image by 90 degrees
rotated_image = resized_image.rotate(-90, expand=True)  # -90 for clockwise, expand=True adjusts the canvas size

# Save and display the rotated image
rotated_image.save("rotated_image.jpg")
print("Rotated image size:", rotated_image.size)
plt.figure(figsize=(8, 8))
plt.title("Rotated Image (90°)")
plt.imshow(rotated_image)
plt.axis('off')
plt.show()
```

Remove the background of the image.only maintain the character.

```python
# Install the rembg library
!pip install onnxruntime
!pip install rembg

from rembg import remove
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# Step 1: Load the image from the URL
url = "https://vignette.wikia.nocookie.net/gameofthrones/images/2/25/707_Daenerys_Profil.jpg/revision/latest?cb=20170829215522&path-prefix=de"
response = requests.get(url)
original_image = Image.open(BytesIO(response.content))

# Display the original image
plt.figure(figsize=(6, 6))
plt.title("Original Image")
plt.imshow(original_image)
plt.axis('off')
plt.show()

# Step 2: Remove the background
image_bytes = BytesIO()
original_image.save(image_bytes, format='PNG')
image_without_bg = remove(image_bytes.getvalue())

# Step 3: Save and display the image without background
output_path = "character_no_bg.png"
with open(output_path, "wb") as f:
    f.write(image_without_bg)

character_image = Image.open(output_path)

# Display the image without background
plt.figure(figsize=(6, 6))
plt.title("Image Without Background")
plt.imshow(character_image)
plt.axis('off')
plt.show()
```

Find 2 creative common images online. Combine these 2 images side by side vertically. Export it as Jpg.

```python
from PIL import Image
import matplotlib.pyplot as plt

# Step 1: Load the images (replace 'image1.jpg' and 'image2.jpg' with your uploaded file names)
image1_path = "Scourge_ROTB.jpg"
image2_path = "Nightbirdigrender[1].jpg"

image1 = Image.open(image1_path)
image2 = Image.open(image2_path)

# Step 2: Resize images to the same width (if necessary)
# This ensures both images align vertically
new_width = min(image1.width, image2.width)
image1 = image1.resize((new_width, int(image1.height * new_width / image1.width)))
image2 = image2.resize((new_width, int(image2.height * new_width / image2.width)))

# Step 3: Combine images vertically
combined_height = image1.height + image2.height
combined_image = Image.new("RGB", (new_width, combined_height))

# Paste images one below the other
combined_image.paste(image1, (0, 0))
combined_image.paste(image2, (0, image1.height))

# Step 4: Convert the combined image to grayscale
gray_combined_image = combined_image.convert("L")  # "L" mode creates grayscale images

# Step 5: Save the grayscale combined image as JPG
output_path = "combined_image_gray.jpg"
gray_combined_image.save(output_path, "JPEG")

# Step 6: Display the grayscale combined image
plt.figure(figsize=(6, 12))
plt.title("Grayscale Combined Image")
plt.imshow(gray_combined_image, cmap="gray")
plt.axis('off')
plt.show()
```
